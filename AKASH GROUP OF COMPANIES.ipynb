```python
#merging the datasets for analysis(fact and customer)
sales1 = fact.merge(customer,how = 'left', on = 'customer_key')
sales1
```

```python
#merging datasets(sales1(fact and customer) with item)
sales2 = sales1.merge(item,how = 'left', on = 'item_key')
sales2
```

```python
#merging datasets(sales2(fact,customer and item) with store)
sales3 = sales2.merge(store,how = 'left', on = 'store_key')
sales3
```

```python
#merging datasets(sales3(fact, customer, item and store) with time)
sales4 = sales3.merge(time,how = 'left', on = 'time_key')
sales4
```

```python
#merging datasets(sales4(fact, customer, item, store and time) with Time)
sales = sales4.merge(Trans,how = 'left', on = 'payment_key')
sales
```
