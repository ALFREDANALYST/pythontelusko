```python
# Convert date column to datetime
companies['Date Joined']= pd.to_datetime(companies['Date Joined'])
companies
```

```python
# Create a new `month` column
companies['year'] = companies['Date Joined'].dt.year
companies.head()
```

```python
# Sample the data
companies_sampled = companies.sample(n=50, random_state=42)

companies_sampled
```

```python
# Calculate years to reach unicorn status
companies['years_to_unicorn'] = companies['year'] - companies['Year Founded']
companies
```

```python
#For each industry, find the longest time to reach unicorn status
longest_by_industry = companies.groupby('Industry')['years_to_unicorn'].max().sort_values(ascending=False)
longest_by_industry
```

```python
# Prepare data for bar()
industries = longest_by_industry.index
years = longest_by_industry.values

# Create the bar plot
plt.figure(figsize=(12, 6))
plt.bar(industries, years, color='steelblue')
plt.xlabel('Industry')
plt.ylabel('Longest Time to Unicorn Status (years)')
plt.title('Longest Time to Reach Unicorn Status by Industry')
plt.xticks(rotation=75, ha='right')
plt.tight_layout()
plt.show()
```
