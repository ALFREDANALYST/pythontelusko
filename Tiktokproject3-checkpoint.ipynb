```python
# Compute the mean `video_view_count` for each group in `verified_status`
### YOUR CODE HERE ###
data.groupby('verified_status')['video_view_count'].mean()
```

```python
#Hypothesis testing
#𝐻0: There is no difference in the average video count between customers who are not verified and customers who are verified.

#𝐻𝐴: There is a difference in the average video count between customers who are not verified and customers who are verified.
```

```python
not_verified = data[data['verified_status'] == 'not verified']
verified = data[data['verified_status'] =='verified']
```

```python
#significance leved=5%,p-value
stats.ttest_ind(a=not_verified['video_view_count'], b=verified['video_view_count'], equal_var=False)
```
