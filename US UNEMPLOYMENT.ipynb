```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

```python
#importing dataset
unemployed = pd.read_csv("E:/PYTHON PROJECTS/unemployed_population_1978-12_to_2023-07.csv")
unemployed
```

```python
#splitting date column
# Convert date column to pandas datetime format
unemployed['date'] = pd.to_datetime(unemployed['date'])

# Split Date into 'date' and 'time' columns
unemployed['year'] = unemployed['date'].dt.year
unemployed['month'] = unemployed['date'].dt.month

# Move 'date' and 'time' columns to the first positions
unemployment = unemployed[['year', 'month'] + ['all','16-24', '25-54', '55-64', '65+', 'less_than_hs', 'high_school', 'some_college', "bachelor's_degree", 'advanced_degree', 'white', 'black', 'hispanic', 'men', 'women']]
unemployment
```
