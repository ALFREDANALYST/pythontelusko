```python
#splitting date column
# Convert Date column to pandas datetime format
tweets['Date'] = pd.to_datetime(tweets['Date'])

# Split Date into 'date' and 'time' columns
tweets['date'] = tweets['Date'].dt.date
tweets['time'] = tweets['Date'].dt.time

# Move 'date' and 'time' columns to the first positions
tweet1 = tweets[['date', 'time'] + ['Target', 'ID', 'flag', 'User', 'Text']]
tweet1
```

```python
tweet = tweet1.set_index('ID')
tweet
```
