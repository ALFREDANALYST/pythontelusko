```python
# while loops
number = 0

while number < 5:
    print(number)
    number = number + 1
```

```python
number = 0

while number < 5:
    print(number)
    if number == 3:
        break
    number = number + 1
```

```python
number = 0

while number < 5:
    print(number)
    if number == 8:
        break
    number = number + 1
else:
    print('no longer < 5')
```

```python
number = 0

while number < 5:
    number = number + 1
    print(number)
    if number == 3:
        continue
        break
    number = number + 1
else:
    print('no longer < 5')
```
