```python
#Import libraries and packages listed above
### YOUR CODE HERE ###
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt

# Load dataset into dataframe
df = pd.read_csv("D:\Downloads\Yellow_Taxi_Trip_Data.csv")
```

```python
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.describe(include='all')
```

```python
# Convert the `date` column to datetime.
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
df.info()
```

```python
#Duration of trips
df['duration'] = df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']
df.head()
```

```python
df.info()
```

```python
df['total_amount'].sort_values(ascending = True).head(30)
```

```python
#Removing outlier amount
df_1 = df[~(df['total_amount'] <= -3.30 )]
df_1.info()
```

```python
df_1['duration'].sort_values(ascending = True).head(5)
```

```python
#Removing oulier durations
df_1['duration'] = pd.to_timedelta(df_1['duration'])

df_2 = df_1[~(df_1['duration'] <= pd.Timedelta(0))]
df_2.info()
```

```python
df_2['duration'].sort_values(ascending=True)
```

```python
df_2['trip_distance'].sort_values(ascending = True).head(50)
```

```python
df_2['trip_distance'].value_counts().sort_values(ascending = False).head(10)
```

```python
#Removing outlier distance
df_clean = df_2[~(df_2['trip_distance'] <= 0.0 )]
df_clean.info()
```

```python
#descriptive stats for clean data
df_clean.describe(include='all')
```

```python
df_clean.shape
```

```python
# Create a month column
df_clean['month'] = df['tpep_pickup_datetime'].dt.month_name()
# Create a day column
df_clean['day'] = df['tpep_pickup_datetime'].dt.day_name()
df_clean.head()
```

```python
#month by trips
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
         'August', 'September', 'October', 'November', 'December']
ax= df_clean['month'].value_counts().plot(kind ='bar')
ax.set_xticklabels(month_order)
```

```python
#day by trips
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
ax = df_clean['day'].value_counts().plot(kind ='bar')
ax.set_xticklabels(day_order)
```

```python
#month by trip distance
ax = df_clean.groupby('month')['trip_distance'].sum().plot(kind='bar')
ax.set_xticklabels(month_order)
```

```python
#day by trip distance
ax = df_clean.groupby('day')['trip_distance'].sum().plot(kind='bar')
ax.set_xticklabels(day_order)
```
