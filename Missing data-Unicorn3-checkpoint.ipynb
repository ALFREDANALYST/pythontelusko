```python
#Create a new column "Year Joined" from "Date Joined".

### YOUR CODE HERE ###
df['Date Joined'] = pd.to_datetime(df['Date Joined'])
df['Year Joined'] = df['Date Joined'].dt.year
df['Month'] = df['Date Joined'].dt.month_name()
df['Weekday'] = df['Date Joined'].dt.day_name()
df.head()
```

```python
import re

def valuation_to_number(val):
    val = val.replace('$', '')
    if 'B' in val:
        return float(val.replace('B', '')) * 1_000_000_000
    elif 'M' in val:
        return float(val.replace('M', '')) * 1_000_000
    else:
        return float(val)

df['Valuation_num'] = df['Valuation'].apply(valuation_to_number)
df.head(10)
```

```python
df.isna().sum()
```

```python
# Create a new df of just the rows that are missing data
df_null = df[pd.isnull(df.City)]
df_null.shape
```

```python
df_null.head()
```

```python
df_null['Country/Region'].value_counts()
```

```python
df['Country/Region'].value_counts()
```

```python
# Define the mapping of country/region to capital city
capital_mapping = {
    'Singapore': 'Singapore',
    'Hong Kong': 'Hong Kong',
    'Bahamas': 'Nassau'
}
mask = df['City'].isna()
df.loc[mask, 'City'] = df.loc[mask, 'Country/Region'].map(capital_mapping)
df.isna().sum()
```

```python
df = df.dropna()
df.shape
```

```python
df.groupby('Country/Region')['Valuation_num'].sum().sort_values(ascending = False)
```

```python
df.groupby('Country/Region')['Valuation'].sum(numeric_only=False).sort_values(ascending = True)
```

```python
df.groupby('Country/Region')['Valuation_num'].sum().sort_values(ascending = False).plot(kind= 'bar')
plt.figure(figsize=(20,5))
plt.xlabel('Country')
plt.ylabel('Valuation')
plt.title('Country by Valuation')
plt.xticks(rotation=75, ha='right')
plt.show()
```

```python
# Plot the sum of valuations per country.

data = df

px.scatter_geo(data, 
               locations='Country/Region', 
               size='Valuation_num', 
               locationmode='country names', 
               color='Country/Region',
               title='Total company valuations by country')
```
