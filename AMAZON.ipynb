```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

```python
#importing dataset
amazon_best = pd.read_csv("E:/GOOGLE3/bestsellers with categories.csv")
amazon_best
```

```python
#exploratory data analysis
amazon_best.info()
```

```python
amazon_best.dtypes
```

```python
amazon_best.isnull().sum()
```

```python
amazon_best.describe(include = "all")
```

```python
amazon_best.drop_duplicates()
```

```python
amazon_best.hist(figsize=(20,30))
```

```python
#genre
genre1 = amazon_best.groupby("Genre")["Genre"].count().sort_values(ascending = False)
genre1
```

```python
genre1.plot.pie(autopct = "%1.1f%%")
plt.title("GENRE DISTRIBUTION")
plt.ylabel('')
```

```python
sns.countplot(y = "Genre", data = amazon_best, palette = 'rocket')
```

```python
sns.countplot(y = 'Year', data = amazon_best, palette = 'rocket')
```

```python
#name
name1 = amazon_best.groupby("Name")['Name'].count().sort_values(ascending = False).head(5)
name1
```

```python
name1.plot(kind = 'bar', title = 'BOOKS BY NUMBER OF TIMES ORDERED', ylabel = 'number of times')
```

```python
#author
author1 = amazon_best.groupby("Author")["Author"].count().sort_values(ascending = False).head(5)
author1
```

```python
author1.plot(kind = 'bar', title = 'AUTHOR BY NUMBER OF TIMES', ylabel = 'number of times')
```

```python
plt.figure(figsize=(12,10))
axis = sns.scatterplot(x = 'Price', y = 'User Rating', data = amazon_best, hue = 'Genre')
plt.title('PRICE VARIES WITH USER RATING')
```

```python
plt.figure(figsize=(12,10))
axis = sns.scatterplot(x = 'Price', y = 'Reviews', data = amazon_best, hue = 'Genre')
plt.title('PRICE VARIATION WITH REVIEWS')
```

```python
plt.figure(figsize=(12,10))
axis = sns.scatterplot(x = 'Reviews', y = 'User Rating', data = amazon_best, hue = 'Genre')
plt.title('REVIEWS VARIATION WITH USER RATING')
```
