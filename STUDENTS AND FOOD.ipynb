```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(color_codes=True)
```

```python
student = pd.read_excel("E:/TABLEAU/student_data.xlsx")
student
```

```python
student.info()
```

```python
student.isnull().sum()
```

```python
student = student.fillna(' ')
student
```

```python
student.isnull().sum()
```

```python
#gender distribution
top_gender = student.groupby('Gender')['Output'].count()
top_gender
```

```python
##plotting gender distribution
top_gender.plot.pie(autopct = '%1.1f%%')
plt.title("GENDER DISTRIBUTION")
plt.ylabel('')
```

```python
#top output
top_output = student.groupby('Output')['Gender'].count()
top_output
```

```python
#top_output plot
top_output.plot.pie(autopct = '%1.1f%%')
plt.title("OUTPUT DISTRIBUTION")
plt.ylabel('')
```

```python
#MARITAL STATUS
marital = student.groupby('Marital status')['Output'].count()
marital
```

```python
marital.plot.pie(autopct = '%1.1f%%')
plt.title("MARITAL STATUS")
plt.ylabel('')
```

```python
#STUDY TIME
study = student.groupby('evening attendance')['Gender'].count()
study
```

```python
study.plot.pie(autopct = '%1.1f%%')
plt.title("STUDY TIME")
plt.ylabel('')
```

```python
#tuition
tuition = student.groupby('Tuition fees up to date')['Gender'].count()
tuition
```

```python
tuition.plot.pie(autopct = '%1.1f%%')
plt.title('TUITION STATUS')
plt.ylabel('')
```

```python
#student debt
debt = student.groupby('Debtor')['Output'].count()
debt
```

```python
debt.plot.pie(autopct = '%1.1f%%')
plt.title("STUDENT DEBTORS")
plt.ylabel('')
```

```python
#education special needs
special_needs  = student.groupby('Educational special needs')['Output'].count()
special_needs
```

```python
special_needs.plot.pie(autopct = '%1.1f%%')
plt.title('STUDENTS SPECIAL NEEDS DISTRIBUTION')
plt.ylabel('')
```

```python
#DISPLACED STUDENTS
displaced = student.groupby('Displaced')['Output'].count()
displaced
```

```python
displaced.plot.pie(autopct = '%1.1f%%')
plt.title("STUDENTS DISPLACED")
plt.ylabel('')
```

```python
food = pd.read_excel("E:/TABLEAU/world_food.xlsx")
food
```

```python
sns.barplot(x='year', y='rice_production', data= food)
plt.figure(figsize=(50,25))
plt.show()
```

```python
rice = food.groupby('year')['rice_production'].sum()
rice
```

```python
rice.plot(kind = 'line', title = 'TOTAL RICE PRODUCTION BY YEAR', ylabel = 'TOTAL RICE PRODUCTION',figsize=(25,15))
```

```python
wheat = food.groupby('year')['wheat_production'].sum()
wheat
```

```python
wheat.plot(kind = 'line', title = 'TOTAL WHEAT PRODUCTION BY YEAR', ylabel = 'TOTAL RICE PRODUCTION', figsize =(25,15))
```

```python
vegetable = food.groupby('year')['vegetable_production'].sum()
vegetable
```
