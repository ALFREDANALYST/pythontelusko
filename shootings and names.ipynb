```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(color_codes=True)
```

```python
shootings = pd.read_excel("E:/TABLEAU/SchoolShootingsUSA.xlsx")
shootings
```

```python
names = pd.read_excel("E:/TABLEAU/TopBabyNamesbyState.xlsx")
names
```

```python
names.info()
```

```python
sns.barplot(x='Gender', y='Occurences', data= names)
plt.show()
```

```python
sns.barplot(x='State', y='Occurences', data= names)
plt.show()
```

```python
names.isnull().sum()
```

```python
#gender distribution
top_gender = names.groupby('Gender')['Occurences'].sum()
top_gender
```

```python
top_gender.plot.pie(autopct = '%1.1f%%')
plt.title("GENDER DISTRIBUTION")
plt.ylabel('')
```

```python
#state distribution
top_state = names.groupby('State')['Occurences'].sum()
top_state
```

```python
#names distribution
top_names = names.groupby('Top Name')['Occurences'].sum()
top_names
```

```python
#top names
most_names = top_names.sort_values(ascending= False).head(10)
most_names
```

```python
most_names.plot(kind = 'bar', title = 'MOST USED NAMES EVER', ylabel = 'NUMBER OF OCCURENCES IN MILLIONS')
```

```python
#LEAST USED NAMES EVER
least_names = top_names.sort_values(ascending= True).head(10)
least_names
```

```python
least_names.plot(kind = 'bar', title = 'LEAST USED NAMES EVER', ylabel = ' NUMBER OF OCCURENCES')
```

```python
#STATE MOST USED NAMES
top_state = names.groupby('State')['Occurences'].sum()
top_state
```

```python
#MOST BIRTHS IN A STATE
most_births = top_state.sort_values(ascending = False).head(10)
most_births
```

```python
most_births.plot(kind = 'bar', title = 'STATE WITH HIGHEST BIRTHS', ylabel = 'TOTAL BIRTHS BY A STATE IN MILLIONS')
```

```python
#DISTRIBUTION BY YEARS
top_year = names.groupby('Year')['Occurences'].sum()
top_year
```

```python
#best year
best_years = top_year.sort_values(ascending = False).head(10)
best_years
```

```python
best_years.plot(kind = 'bar', title = 'YEAR WITH MOST NAME OCCURENCES', ylabel = 'NUMBER OF OCCURENCES')
```

```python
#LEAST YEARS
least_years = top_year.sort_values(ascending = True).head(10)
least_years
```

```python
least_years.plot(kind = 'bar', title = 'LEAST YEARS OF OCCURENCES', ylabel = 'NUMBER OF OCCURENCES')
```

```python
#LEAST BIRTHS IN A STATE
least_births = top_state.sort_values(ascending = True).head(10)
least_births
```

```python
least_births.plot(kind = 'bar', title = 'LEAST OCCURENCES OF NAMES BY STATE', ylabel = 'NUMBER OF OCCURENCES')
```

```python
top_year.plot(kind = 'line', title = 'TOTAL NAME OCCURENCES BY YEAR', ylabel = 'TOTAL NAME OCCURENCES')
```
